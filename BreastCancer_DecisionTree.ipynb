```python
import pickle
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("BreastCancerDataset.csv")
df.head(7)
```

```python
df.columns
```

```python
df.isnull().sum()  # Checking null values #No Null Values in Dataset.
```

```python
X = df.drop("diagnosis",axis=1)     # Extract the features (all columns except Attritions) from the dataset
le = LabelEncoder()
Y = le.fit_transform(df['diagnosis'])
Y
```

```python
# Spliting the data into train and test
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

Loading Preprocessor

```python
# Load a preprocessor object from a pickled file
with open("/content/BCPreProprocessing.pkl","rb") as f:
    preprocessor=pickle.load(f)
```

```python
preprocessor
```

Transforming the data

```python
# Transform the training data using the preprocessor object or PipeLine
Proprocessed_X_train = preprocessor.fit_transform(X_train)
Preprocessed_X_test=preprocessor.transform(X_test)
```

```python
Y_train[57]
```

Balancing The Data

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_train_smote,Y_train_smote=smote.fit_resample(Proprocessed_X_train,Y_train)
```

```python
params = {
    "criterion":("gini", "entropy"), #quality of split
    "splitter":("best", "random"), # searches the features for a split
    "max_depth":(list(range(1, 10))), #depth of tree range from 1 to 9
    "min_samples_split":[2, 3, 4,5,6,7],    #the minimum number of samples required to split internal node
    "min_samples_leaf":list(range(1, 10)),#minimum number of samples required to be at a leaf node,we are passing list which is range from 1 to 9
}
```

```python
from sklearn.model_selection import GridSearchCV
tree_clf=DecisionTreeClassifier()
tree_cv=GridSearchCV(tree_clf,params,scoring="f1",n_jobs=-1,verbose=2,cv=5)
```

```python
Y_Pred=tree_cv.predict(Preprocessed_X_test)
print(classification_report(Y_test,Y_Pred))
```
